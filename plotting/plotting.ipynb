```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

# Set Seaborn style to dark
sns.set_theme(style="darkgrid", context="notebook")

# Data for multiagent_llama3.1 and multiagent_llama3.2 for 2 rounds
data_combined_2rounds = {
    'Model': ['3.1', '3.1', '3.1', '3.2', '3.2', '3.2'],
    'Agents': [2, 3, 4, 2, 3, 4],
    'Precision': [0.4322, 0.4261, 0.4027, 0.2800, 0.2403, 0.2799],
    'Recall':    [0.3591, 0.3829, 0.3916, 0.3293, 0.3299, 0.3516],
    'F1':        [0.3959, 0.4043, 0.3972, 0.3049, 0.2850, 0.3158]
}

df_combined_2rounds = pd.DataFrame(data_combined_2rounds)

# Plot metrics for 2 rounds
metrics = ['Precision', 'Recall', 'F1']

for metric in metrics:
    plt.figure(figsize=(6, 4))
    
    for model in df_combined_2rounds['Model'].unique():
        subset = df_combined_2rounds[df_combined_2rounds['Model'] == model]
        plt.plot(
            subset['Agents'], 
            subset[metric], 
            marker='o', 
            label=f'Llama {model}'
        )
    
    plt.title(f'{metric} vs. Number of Agents (2 Rounds)')
    plt.xlabel('Number of Agents')
    plt.ylabel(metric)
    plt.grid(True)
    plt.xticks(df_combined_2rounds['Agents'].unique())
    plt.legend(title='Model')
    plt.tight_layout()
    plt.show()
```

```python
# import matplotlib.pyplot as plt


data_ablation_rounds = {
    'Model': ['3.1', '3.1', '3.2', '3.2'],
    'Rounds': [2, 3, 2, 3],
    'Precision': [0.4322, 0.4134, 0.2800, 0.2830],
    'Recall':    [0.3591, 0.3654, 0.3293, 0.3134],
    'F1':        [0.3959, 0.3893, 0.3049, 0.2984]
}

df_ablation_rounds = pd.DataFrame(data_ablation_rounds)


# Create subplots for the three metrics
fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharex=True, sharey=False)

# Iterate through metrics and plot in subplots
for i, metric in enumerate(metrics):
    for model in df_ablation_rounds['Model'].unique():
        subset = df_ablation_rounds[df_ablation_rounds['Model'] == model]
        axs[i].plot(
            subset['Rounds'], 
            subset[metric], 
            marker='o', 
            label=f'Llama {model}'
        )
    axs[i].set_title(f'{metric} vs. Number of Rounds (Fixed: 2 Agents)')
    axs[i].set_xlabel('Number of Rounds')
    axs[i].set_ylabel(metric)
    axs[i].grid(True)
    axs[i].set_xticks(df_ablation_rounds['Rounds'].unique())

# Add legend to the first subplot
axs[0].legend(title='Model', loc='upper right')

plt.tight_layout()
plt.show()
```
